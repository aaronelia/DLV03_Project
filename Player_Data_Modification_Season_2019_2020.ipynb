## Defensive Actions

```python
defensive_actions = pd.read_csv(
    r"C:\Users\aaron\OneDrive\HSLU\Hackdays\FBREF_Player_Data\defensive_actions_2019_2020.csv")
defensive_actions= defensive_actions.drop("Unnamed: 0",1)
defensive_actions
```

```python
defensive_actions=defensive_actions.rename(columns={"TklW":"Tackles","Tkl%":"Tack/DP%","Past":"Dribbled Past","Press":"Pressures","Succ":"Pressure Regains","Att 3rd.1":"Pressures in Opposing Half","Sh":"Blocks/Shot","Int":"Interceptions","Tkl+Int":"Tack&Int","Clr":"Clearances","Err":"Errors","%":"Press Succ%"})
defensive_actions
```

```python
defensive_actions["Dribbles Stopped"]=defensive_actions["Tkl.1"] /(defensive_actions["Dribbled Past"]*defensive_actions["Tkl.1"])*100
defensive_actions["Dribbles Stopped"] = defensive_actions["Dribbles Stopped"].fillna(0)
```

```python
defensive_actions=defensive_actions.drop(columns=["Rk","Age","Born","90s","Tkl","Def 3rd","Mid 3rd","Att 3rd","Tkl.1","Att","Def 3rd.1","Mid 3rd.1","Blocks","ShSv","Pass"])
defensive_actions
```

```python
defensive_actions.to_csv("df_defensive_actions.csv", index=False, encoding="utf-8")
```
